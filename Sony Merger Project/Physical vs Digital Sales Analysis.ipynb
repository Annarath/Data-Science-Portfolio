```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

```python
df = pd.read_csv("/Users/unun/Desktop/Sony Project/Digital Vs Physical Music .csv")
```

```python
df.set_index('Types', inplace=True)
```

```python
df
```

```python
df['2010'] = df['2010'].str.replace(',','')
df['2011'] = df['2011'].str.replace(',','')
df['2012'] = df['2012'].str.replace(',','')
df['2013'] = df['2013'].str.replace(',','')
df['2014'] = df['2014'].str.replace(',','')
```

```python
df['2015'] = df['2015'].str.replace(',','')
df['2016'] = df['2016'].str.replace(',','')
df['2017'] = df['2017'].str.replace(',','')
df['2018'] = df['2018'].str.replace(',','')
df['2019'] = df['2019'].str.replace(',','')
df['2020'] = df['2020'].str.replace(',','')
```

```python
df['2010'] = pd.to_numeric(df['2010'])
df['2011'] = pd.to_numeric(df['2011'])
df['2012'] = pd.to_numeric(df['2012'])
df['2013'] = pd.to_numeric(df['2013'])
df['2014'] = pd.to_numeric(df['2014'])
df['2015'] = pd.to_numeric(df['2015'])
df['2016'] = pd.to_numeric(df['2016'])
df['2017'] = pd.to_numeric(df['2017'])
df['2018'] = pd.to_numeric(df['2018'])
df['2019'] = pd.to_numeric(df['2019'])
df['2020'] = pd.to_numeric(df['2020'])
```

```python
df.dtypes
```

```python
plt.plot(df.loc['Physical'])
```

```python
x1 = df.loc['Physical']
x2 = df.loc['Digital']
fig = plt.figure(figsize=(10,6), dpi=100)

ax=fig.add_axes([0,0,1,1])
ax.set_title('Physical VS Digital Sales of Whole Industry')
ax.plot(x1, label='Physical Sales')
ax.plot(x2, label='Digital Sales')
ax.legend(loc=0)
plt.xlabel('Year')
plt.ylabel('Sales Revenue')
```
